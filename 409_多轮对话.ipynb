```python
# ä» openai åº“ä¸­å¯¼å…¥ OpenAI ç±»ï¼Œç”¨äºè°ƒç”¨ DeepSeek çš„ Chat API
from openai import OpenAI  
# å¯¼å…¥ os æ¨¡å—ï¼Œç”¨äºè®¿é—®ç³»ç»Ÿç¯å¢ƒå˜é‡
import os  

# ä»ç³»ç»Ÿç¯å¢ƒå˜é‡ä¸­è·å–åä¸º DEEPSEEK_API_KEY çš„ API å¯†é’¥
api_key = os.environ['DEEPSEEK_API_KEY']  

# åˆ›å»ºä¸€ä¸ª OpenAI å®¢æˆ·ç«¯å¯¹è±¡ï¼ŒæŒ‡å®šä½¿ç”¨ DeepSeek çš„ API ç½‘å…³åœ°å€ï¼Œå¹¶ä¼ å…¥ API å¯†é’¥è¿›è¡Œé‰´æƒ
client = OpenAI(base_url='https://api.deepseek.com', api_key=api_key)

# æ„å»ºåˆå§‹å¯¹è¯æ¶ˆæ¯ï¼Œè¡¨ç¤ºç”¨æˆ·å‘æ¨¡å‹æå‡ºç¬¬ä¸€ä¸ªé—®é¢˜
messages = [{'role': 'user', 'content': '1+1ç­‰äºå‡ '}]

# å‘ DeepSeek æ¨¡å‹å‘é€èŠå¤©è¯·æ±‚ï¼Œä¼ å…¥æ¨¡å‹åå’Œæ¶ˆæ¯å†å²
response = client.chat.completions.create(
    model='deepseek-chat',  # æŒ‡å®šè¦ä½¿ç”¨çš„ DeepSeek èŠå¤©æ¨¡å‹
    messages=messages       # å‘é€çš„æ¶ˆæ¯åˆ—è¡¨ï¼ˆå¯¹è¯ä¸Šä¸‹æ–‡ï¼‰
)

# å°†æ¨¡å‹çš„å›å¤ï¼ˆAI å›å¤ï¼‰åŠ å…¥åˆ°æ¶ˆæ¯åˆ—è¡¨ä¸­ï¼Œä¾¿äºç»´æŒå¯¹è¯ä¸Šä¸‹æ–‡
messages.append(response.choices[0].message)

# æ‰“å°æ¨¡å‹çš„ç¬¬ä¸€è½®å›å¤å†…å®¹
print(f'Messages Round 1: {response.choices[0].message}')

# ç”¨æˆ·ç»§ç»­æé—®â€œå†åŠ 1å‘¢â€ï¼Œå°†æ­¤é—®é¢˜åŠ å…¥æ¶ˆæ¯åˆ—è¡¨ï¼Œç»´æŒä¸Šä¸‹æ–‡
messages.append({'role': 'user', 'content': 'å†åŠ 1å‘¢'})

# å†æ¬¡å‘æ¨¡å‹å‘é€å®Œæ•´çš„å¯¹è¯å†å²ï¼Œä»¥è·å–æ–°ä¸€è½®å›å¤
response = client.chat.completions.create(
    model='deepseek-chat',  # ä½¿ç”¨ç›¸åŒçš„ DeepSeek æ¨¡å‹
    messages=messages        # å‘é€åŒ…å«ä¸¤è½®å¯¹è¯çš„æ¶ˆæ¯åˆ—è¡¨
)

# å°†æ¨¡å‹çš„ç¬¬äºŒæ¬¡å›å¤åŠ å…¥æ¶ˆæ¯åˆ—è¡¨
messages.append(response.choices[0].message)

# æ‰“å°æ¨¡å‹çš„ç¬¬äºŒè½®å›å¤å†…å®¹
print(f'Messages Round 2:{response.choices[0].message}')
```
