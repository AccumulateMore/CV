# 1. 注意力分数

```python
import math
import torch
from torch import nn
from d2l import torch as d2l
```

```python
# 遮蔽softmax操作
# 也即只针对句子的前valid_lens长度的单词来做softmax，后面的单词都忽略掉

def masked_softmax(X, valid_lens):
    """通过在最后一个轴上遮蔽元素来执行softmax操作"""
    
    # 如果valid_lens，那么不需要做掩码，就是正常的softmax就可以了
    if valid_lens is None:
        return nn.functional.softmax(X, dim=-1)
    else:
        shape = X.shape
        if valid_lens.dim() == 1:
            # 将valid_lens重复shape[1]遍
            valid_lens = torch.repeat_interleave(valid_lens, shape[1])
        else:
            valid_lens = valid_lens.reshape(-1)
        
        # 用很小的值 value = -1e6 来填充后面的部分（因为e^(-1e6)几乎为0了）
        X = d2l.sequence_mask(X.reshape(-1, shape[-1]), valid_lens, value=-1e6)
        
        return nn.functional.softmax(X.reshape(shape), dim=-1)
```

```python
# 演示此函数是如何工作
X = torch.rand(2,2,4)
print(X)
print()

output = masked_softmax(X, torch.tensor([2,3]))
print(output)
```

```python
X = torch.rand(2,2,4)
print(X)
print()

output = masked_softmax(X, torch.tensor([[1,3],[2,4]]))
print(output)
```

```python
# 加性注意力
# 这里是向量版本的query、key和value
class AdditiveAttention(nn.Module):
    """加性注意力"""
    
    # 有W_k、W_q和w_v三个可以学习的参数
    def __init__(self, key_size, query_size, num_hiddens, dropout, **kwargs):
        super(AdditiveAttention, self).__init__(**kwargs)
        self.W_k = nn.Linear(key_size, num_hiddens, bias=False)
        self.W_q = nn.Linear(query_size, num_hiddens, bias=False)
        self.w_v = nn.Linear(num_hiddens, 1, bias=False)
        self.dropout = nn.Dropout(dropout)
        
    # valid_lens：表示针对每个query，应该考虑前多少个key-value对
    # 因为有一部分key-value对可能是pad出来的，所以这里是用来屏蔽被pad的这些部分的
    def forward(self, queries, keys, values, valid_lens):
        # 做乘法后的queries和keys
        queries = self.W_q(queries)  # queries的维度为"batch_size * number of queries * h"
        keys = self.W_k(keys)  # 而keys的维度为"batch_size * number of keys * h"
        
        print(f"queries' shape is {queries.shape}")
        print(f"keys' shape is {keys.shape}")
        # print(f"queries is {queries}")
        # print(f"keys is {keys}")
        
        # 因为这里queries和keys的中间维度不同，不能直接相加；所以在这里要做一个广播
        # unsqueeze之后，queries维度变成"batch_size * number of queries * 1 * h"
        # 而keys的维度则变成了"batch_size * 1 * number of keys * h"
        # 从而"bs * nq * 1 * h"和"bs * 1 * nk * h"就可以利用广播机制相加了，得到的结果为"bs * nq * nk * h"
        features = queries.unsqueeze(2) + keys.unsqueeze(1)
        features = torch.tanh(features)  # 对每个值都做一个tanh
        
        print(f"features' shape is {features.shape}")
        
        # 然后再和w_v做矩阵乘法，会得到"bs * nq * nk * 1"的结果；所以再用squeeze(-1)去掉最后一维
        scores = self.w_v(features).squeeze(-1)
        
        print(f"scores' shape is {scores.shape}")
        print(f"scores is {scores}")
        
        # 然后再对每个注意力分数再做一次masked_softmax，过滤掉不需要的key-value对们
        self.attention_weights = masked_softmax(scores, valid_lens)
        
        print(f"self.attention_weights' shape is {self.attention_weights.shape}")
        print(f"self.attention_weights is {self.attention_weights}")
        
        # 再是先对权重做一个dropout，再将这个注意力权重和values放在一起做batch的矩阵乘法，得到结果
        # 返回结果output的维度为"batch_size * number of queries * dim(v)"
        # 也即“批量数 * q的个数 * v的长度”
        output = torch.bmm(self.dropout(self.attention_weights), values)
        
        print(f"output's shape is {output.shape}")
        
        return output
```

```python
# 演示上面的AdditiveAttention类

# torch.normal函数用法：https://blog.csdn.net/qq_34690929/article/details/80029974
queries = torch.normal(0, 1, (2,1,20))  # 表示生成的queries的batch_size为2，每个batch包含1个query，每个query的长度为20
keys = torch.ones((2,10,2))  # 表示生成的keys的batch_size为2，每个batch包含10个key，每个key的长度为2

# 将1-40共40个数字分到1*10*4的tensor中，然后再把第一维repeat一遍，所以values的维度为2*10*4
# 表示values的batch_size为2，每个batch包含10个value，每个value的长度为4
values = torch.arange(40, dtype=torch.float32).reshape(1,10,4).repeat(2,1,1)  
valid_lens = torch.tensor([2,6])

print(f"queries is {queries}")
# print(f"keys is {keys}")
# print(f"values is {values}")
print("-"*30)

attention = AdditiveAttention(key_size=2, query_size=20, num_hiddens=8, dropout=0.1)  
attention.eval()

output = attention(queries, keys, values, valid_lens)

print("-"*30)
print(output)  # 返回结果为“批量数 * q的个数 * v的长度”
```

```python
# 注意力权重
d2l.show_heatmaps(attention.attention_weights.reshape((1,1,2,10)),
                 xlabel='Keys', ylabel='Queries')
```

```python
# 缩放点积注意力
# 在query和key的维度相同时方可使用

class DotProductAttention(nn.Module):
    """缩放点积注意力"""
    
    def __init__(self, dropout, **kwargs):
        super(DotProductAttention, self).__init__(**kwargs)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, queries, keys, values, valid_lens=None):
        # 获取query或key的维度d
        d = queries.shape[-1]
        
        # 这里keys要做一个转置（因为计算公式即为"Q*K^T / sqrt(d)"）
        scores = torch.bmm(queries, keys.transpose(1,2)) / math.sqrt(d)
        
        self.attention_weights = masked_softmax(scores, valid_lens)
        
        output = torch.bmm(self.dropout(self.attention_weights), values)
        
        # 返回的结果应该也是“批量数 * q的个数 * v的长度”？？？
        return output
```

```python
# 演示上述的DotProductAttention类

queries = torch.normal(0, 1, (2,1,2))

attention = DotProductAttention(dropout=0.5)
attention.eval()

attention(queries, keys, values, valid_lens)
```

```python
# 均匀的注意力权重
d2l.show_heatmaps(attention.attention_weights.reshape((1,1,2,10)),
                 xlabel='Keys', ylabel='Queries')
```
