# 1. 注意力机制
### 注意力汇聚：Nadaraya-Watson核回归

```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
# 生成数据集

n_train = 50


# 给出的真实函数
def f(x):
    return 2 * torch.sin(x) + x**0.8

# torch.sort解析：https://hxhen.com/archives/226
# 默认按照dim=1按行排序，默认descending=False从小到大排序
x_train, _ = torch.sort(torch.rand(n_train) * 5)  # 生成50个随机的[0, 5)之间的数字，并升序排序，作为train数据
y_train = f(x_train) + torch.normal(0.0, 0.5, (n_train,))  # 训练数据上面加上噪音

x_test = torch.arange(0, 5, 0.1)  # 生成连续数组作为test数据
y_truth = f(x_test)

n_test = len(x_test)
n_test
```

```python
def plot_kernel_reg(y_hat):
    d2l.plot(x_test, [y_truth, y_hat], 'x', 'y', legend=['Truth','Pred'],
             xlim=[0,5], ylim=[-1,5])
    
    d2l.plt.plot(x_train, y_train, 'o', alpha=0.5)

# 最简单的预测：直接求平均y_train.mean()
# 所以画出来是一条紫色的虚线
# ps：repeat_interleave函数解析：https://blog.csdn.net/weixin_45261707/article/details/119187799
# 所以这里就是将这份平均值复制了50遍，作为最后的输出，所以才是一条水平线
y_hat = torch.repeat_interleave(y_train.mean(), n_test)
plot_kernel_reg(y_hat)


# 黄色的点是随机的那些x_train和y_train的点们
# 蓝色的线是y_truth画出的线
# 紫色的线上y_hat（预测值）画出的线
```

```python
# 非参数注意力汇聚（原翻译：非参数注意力池化，并不好理解）

# 此时X_repeat的大小变成了n_test * n_train
# 也就是把x_test重复了50遍的结果
X_repeat = x_test.repeat_interleave(n_train).reshape((-1, n_train))
print(x_train.shape)
print(x_test.shape)
print(X_repeat.shape)

# 使用Nadaraya-Watson核回归方法 + 高斯核函数来获得注意力权重
# nn.functional.softmax解析：https://www.cnblogs.com/wanghui-garcia/p/10675588.html
# 另一个解析为：https://blog.csdn.net/Will_Ye/article/details/104994504
# 相当于给更近的点以更大的权重
attention_weights = nn.functional.softmax(-(X_repeat - x_train)**2 / 2, dim=1)
print(attention_weights.shape)

# 相当于预测x_test中的每个点时，都看到了x_train中所有的50个点；但是看到的每个点的权重不同，近的点会有更高的权重
y_hat = torch.matmul(attention_weights, y_train)
print(y_train.shape)
print(y_hat.shape)

plot_kernel_reg(y_hat)
```

```python
# 注意力权重

# 这个图是逐行逐行看的
# attention_weights是50*50，
# 也即对于x_test中的每个点，画出他看到的x_train中所有的50个点的不同点的权重值
d2l.show_heatmaps(attention_weights.unsqueeze(0).unsqueeze(0),
                  xlabel='Sorted training inputs', ylabel='Sorted test inputs')   
```

```python
# 带参数注意力汇聚

# 先介绍torch.bmm乘法（带batch的矩阵乘法）
# 假定两个张量的形状分别是(n,a,b)和(n,b,c)，它们的批量矩阵乘法输出的形状为(n.a,c)  
X = torch.ones((2,1,4))
Y = torch.ones((2,4,6))
torch.bmm(X, Y).shape
```

```python
# 使用小批量矩阵乘法来计算小批量数据中的加权平均值

weights = torch.ones((2,10)) * 0.1
values = torch.arange(20.0).reshape((2,10))
torch.bmm(weights.unsqueeze(1), values.unsqueeze(-1))
```

```python
# 带参数的注意力汇聚
# 就是在非参数的基础上，再加一个w参数进来
class NWKernelRegression(nn.Module):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = nn.Parameter(torch.rand((1,),requires_grad=True))  
        
    def forward(self, queries, keys, values):
        queries = queries.repeat_interleave(keys.shape[1]).reshape(-1,keys.shape[1])   
        self.attention_weights = nn.functional.softmax(-((queries - keys) * self.w)**2/2,dim=1)     
        return torch.bmm(self.attention_weights.unsqueeze(1),values.unsqueeze(-1)).reshape(-1)   
```

```python
# 将训练数据集转换为键和值
X_tile = x_train.repeat((n_train, 1))
Y_tile = y_train.repeat((n_train, 1))
keys = X_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape((n_train,-1))  
values = Y_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape(n_train, -1) 
```

```python
# 训练带参数的注意力汇聚模型
net = NWKernelRegression()
loss = nn.MSELoss(reduction='none')
trainer = torch.optim.SGD(net.parameters(), lr=0.5)
animator = d2l.Animator(xlabel='epoch',ylabel='loss',xlim=[1,5])

for epoch in range(5):
    trainer.zero_grad()
    l = loss(net(x_train, keys, values), y_train) / 2
    l.sum().backward()
    trainer.step()
    print(f'epoch {epoch+1}, loss {float(l.sum()):.6f}')
    animator.add(epoch+1, float(l.sum()))
```

```python
# 预测结果绘制
keys = x_train.repeat((n_test, 1))
values = y_train.repeat((n_test, 1))
y_hat = net(x_test, keys, values).unsqueeze(1).detach()
plot_kernel_reg(y_hat)
```

```python
# 曲线在注意力权重较大的区域变得更不平滑
d2l.show_heatmaps(net.attention_weights.unsqueeze(0).unsqueeze(0),
                 xlabel='Sorted training inputs', ylabel='Sorted testing inputs')   
```
