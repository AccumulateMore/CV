# 2. Tensorboard 写日志

```python
from torch.utils.tensorboard import SummaryWriter
writer = SummaryWriter("logs")  # 创建一个logs文件夹，writer写的文件都在该文件夹下
# writer.add_image()
for i in range(100):
    writer.add_scalar(tag="y=2x", scalar_value=2*i, global_step=i)
writer.close()
```

# 4. Tensorboard 读图片

```python
from torch.utils.tensorboard import SummaryWriter
from PIL import Image
import numpy as np

img_path1 = "Data/hymenoptera_data/train/ants/0013035.jpg" 
img_PIL1 = Image.open(img_path1)
img_array1 = np.array(img_PIL1)
print(img_PIL1)
print(img_array1.shape)


img_path2 = "Data/hymenoptera_data/train/bees/90179376_abc234e5f4.jpg"
img_PIL2 = Image.open(img_path2)
img_array2 = np.array(img_PIL2)
print(img_PIL2)
print(img_array2.shape)


writer = SummaryWriter("logs")
# gloabl_step=1:表示该图片在第1步; dataformats="HWC"表示当前图片为HWC格式
writer.add_image(tag="test", img_tensor=img_array1, global_step=1, dataformats="HWC")
# global_step=2:表示该图片在第2步
writer.add_image("test", img_array2, 2, dataformats="HWC")
writer.close()
```
