# 3. 网络模型添加

我们在现有的vgg16模型基础之上，再使用add_module()方法，来增加一个nn.Linear(1000,10)线性层，使得其能够用于训练CIFAR10数据集。

```python
import torchvision
from torch import nn

dataset = torchvision.datasets.CIFAR10("./dataset",train=True,transform=torchvision.transforms.ToTensor(),download=True)

# 下载卷积层对应的参数是多少、池化层对应的参数时多少，这些参数时ImageNet训练好了的
vgg16_true = torchvision.models.vgg16(pretrained=True)
# 在VGG16后面添加一个线性层，使得输出为适应CIFAR10的输出，CIFAR10需要输出10个种类
vgg16_true.add_module('add_linear', nn.Linear(1000, 10))

print(vgg16_true)
```

# 4. 网络模型修改

Q：如何在网络模型的中间，而不是结尾部分增加新的层呢？
A：可以直接访问某一层并修改，例如：vgg16_false.classifier[6] = nn.Linear(4096, 10)

```python
import torchvision
from torch import nn

vgg16_false = torchvision.models.vgg16(pretrained=False)  # 没有预训练的参数     
print(vgg16_false)

vgg16_false.classifier[6] = nn.Linear(4096, 10)
print(vgg16_false)
```
